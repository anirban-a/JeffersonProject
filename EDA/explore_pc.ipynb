```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
pd.set_option("max_rows", 25)
```

```python
# Apply the default theme
sns.set_theme()
sns.set_style("whitegrid")
```

```python
data_path = '/home/achara/data/achara/'
```

```python
df = pd.read_csv(data_path + 'pc_data_2017_to_2019_2.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(
    data=df.groupby(['month'], as_index=False)['PC_Water_EXO'].mean(),
    x='month',
    y='PC_Water_EXO'
)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(
    data=df.groupby(['month'], as_index=False)['PRECTOTCORR'].mean(),
    x='month',
    y='PRECTOTCORR'
)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10,10))
sns.barplot(
    data=df.groupby(['month'], as_index=False)['Temperature_Water_EXO'].mean(),
    x='month',
    y='Temperature_Water_EXO'
)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.regplot(
    data=df.groupby(['month'], as_index=False)['PC_Water_EXO'].mean(),
    x='month',
    y='PC_Water_EXO',
    order=1
)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(
    data=df.groupby(['month'], as_index=False)['NH4'].mean(),
    x='month',
    y='NH4'
)
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.regplot(data=df.groupby(['SensorDepth_Water_EXO'], as_index=False)['PC_Water_EXO'].mean(),
    x='SensorDepth_Water_EXO',
    y='PC_Water_EXO', order=1, line_kws={'color':'red'},)
plt.show()
```
